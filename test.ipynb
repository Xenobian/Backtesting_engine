```python
import pandas as pd
```

```python
df = pd.read_csv('backtest.csv')
```

```python
df['pl'] = df['pl'] * 25

df['pl_after_cost'] = df['pl'] - (df['trades'] * 100)

df['cumilativePL'] = df['pl'].cumsum()



df.to_csv('stats.csv')
```

```python
negative_days = len(df[df['pl_after_cost'] < 0])
positive_days = len(df[df['pl_after_cost'] > 0])
print(positive_days,negative_days)
```

```python
import matplotlib.pyplot as plt

ax = plt.gca()


df.plot(kind='line',x='date',y='cumilativePL',ax=ax)
```
